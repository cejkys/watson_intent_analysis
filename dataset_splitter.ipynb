```python
import pandas as pd
from sklearn.utils import shuffle
```

```python
colnames = ['Query', 'Category']
df = pd.read_csv('query_2.csv', sep=',', names=colnames, header=None)
print(df.shape)
df.head(2)
```

```python
cf = df['Category'].value_counts().reset_index()
cf = cf.rename(columns={'Category': 'Count', 'index': 'Category'})
print(cf.shape)
cf.head(2)
```

```python
cross_1 = shuffle(df)
cross_2 = shuffle(df)
cross_3 = shuffle(df)
cross_4 = shuffle(df)
cross_5 = shuffle(df)
```

```python
train = pd.DataFrame()
test = pd.DataFrame()
for i in range(cf.shape[0]):
    split_number = int(round(cf['Count'][i]*0.8))
    category_examples = cross_5[cross_5['Category'].str.match(cf['Category'][i])].reset_index()
    counter = 0
    for index, row in category_examples.iterrows():
        if counter <= split_number:
            train = train.append(row)
        else:
            test = test.append(row)
        counter = counter + 1
```

```python
train = train.drop(columns=['index']).reset_index(drop=True)
train['Query'] = train['Query'].astype(str)
train['Category'] = train['Category'].str.replace(' ', '_')
```

```python
test.to_csv('./testset_5.csv')
train.to_csv('./train_set_5.csv', index=False)
```
